```python
################################################## Coding Ninjas ##################################################
def printPascal(n:int):
    # Write your code here.
    # Return a list of lists.
    x_old = []
    final = []
    for i in range(n):
        x = []
        x.append(1)
        for j in range(i-1):          
            x.append(x_old[j]+x_old[j+1])
        if i!=0:
            x.append(1)
        x_old = x
        final.append(x)
    return(final)
        
################################################## LeetCode ##################################################
class Solution(object):
    def generate(self, numRows):
        """
        :type numRows: int
        :rtype: List[List[int]]
        """
        x_old = []
        final = []
        for i in range(numRows):
            x = []
            x.append(1)
            for j in range(i-1):          
                x.append(x_old[j]+x_old[j+1])
            if i!=0:
                x.append(1)
            x_old = x
            final.append(x)
        return(final)
```
