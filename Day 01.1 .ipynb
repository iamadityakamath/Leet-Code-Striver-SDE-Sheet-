# 2) Create a new sample matrix

new matrix of same size and assign all values as false.

while iterating through original matrix, if it is false, assign true to the new matrix. 

Then in end, iterate through new matrix and where ever its True, convert to 0 in original matrix.


```python
################################################## Coding Ninjas ##################################################
def setZeros(matrix: List[List[int]]) -> None: 
    # Storing dimensions of matrix in n and m.
    n = len(matrix)
    m = len(matrix[0])
    isZero = [[False for _ in range(m)] for _ in range(n)]
    
    for i in range(n):
        for j in range(m):
            if matrix[i][j] == 0:
                for x in range(n): isZero[x][j]= True
                for x in range(m): isZero[i][x]= True
     
    for i in range(n):
        for j in range(m):
            if isZero[i][j]:
                matrix[i][j]=0

################################################## LeetCode ##################################################
def setZeroes(self, matrix):        
# Storing dimensions of matrix in n and m.
    n = len(matrix)
    m = len(matrix[0])
    isZero = [[False for _ in range(m)] for _ in range(n)]

    for i in range(n):
        for j in range(m):
            if matrix[i][j] == 0:
                for x in range(n): isZero[x][j]= True
                for x in range(m): isZero[i][x]= True

    for i in range(n):
        for j in range(m):
            if isZero[i][j]:
                matrix[i][j]=0
```
